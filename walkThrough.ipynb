# Evaluation Metrics
metric.csv file shows the evaluation metrics of the run.

    1. accuracy: the overall accuracy for all test samples(merging all unseen types as a single "others" class)
    2. high_confidence_accuracy: accuracy for only samples with high confidence(excluding the "other" class)
    3. precision_unseen: precision for predicting the class "others"
    4. recall_unseen: recall for predicting the class "others"
    5. f1_unseen: f1 score for predicting the class "others"

```python
pd.read_csv('outputs/metrics.csv')
```
